```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import PatternFill, Font

# -----------------------------
# pandas：Excelファイルの読み込み
# -----------------------------
df_2022 = pd.read_excel("2022_年間売上表.xlsx")
df_2023 = pd.read_excel("2023_年間売上表.xlsx")

# -----------------------------
# データの連結と集計
# -----------------------------
df_all = pd.concat([df_2022, df_2023], ignore_index=True)

# 商品・売上年ごとに金額を合計
df_group = df_all.groupby(["商品", "売上年"], as_index=False)["金額（千円）"].sum()

# -----------------------------
# pandas：Excelファイルに書き込み
# -----------------------------
output_file = "売上集計表.xlsx"
df_group.to_excel(output_file, index=False)

# -----------------------------
# openpyxl：ヘッダーの書式設定
# -----------------------------
wb = load_workbook(output_file)
ws = wb.active

# ヘッダー行（1行目）を薄いグレーに設定
fill = PatternFill(start_color="F2F2F2", end_color="F2F2F2", fill_type="solid")

for cell in ws[1]:
    cell.fill = fill
    cell.font = Font(bold=True)  # 太字（指定はないが参照は太くなっている）

wb.save(output_file)
```
